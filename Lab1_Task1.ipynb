```python
import numpy as np

# Task 1.1: Create a random 4x4 array from exponential distribution

print("=== Task 1.1: Creating 4x4 Exponential Distribution Array ===")

# Create a 4x4 array with exponential distribution
# scale=1.0 is the parameter (lambda = 1/scale)
exp_array = np.random.exponential(scale=1.0, size=(4, 4))

print("4x4 Exponential Distribution Array:")
print(exp_array)
```

```python
# Task 1.2: Create 100000x1 array and visualize distributions

print("=== Task 1.2: Distribution Visualization ===")

import matplotlib.pyplot as plt

# Generate arrays
exp_data = np.random.exponential(scale=1.0, size=(100000,1))
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plot histograms
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")

# Adjust plot
plt.axis([0, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# Task 1.3: 3D plot of Z = X**2 + Y**2

print("=== Task 1.3: 3D Surface Plot of Z = X² + Y² ===")

# Task 1.3: Plot Z = X**2 + Y**2 in 3D, limit x and y to [-5, 5]

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z = X**2 + Y**2
Z = X**2 + Y**2

# Create 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
# Task 1.4: Calculate pearson and spearman correlation for Pokemon data

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Download and load Pokemon dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select required features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
pokemon_stats = df[features]

# Calculate Pearson correlation coefficient
pearson_corr = pokemon_stats.corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = pokemon_stats.corr(method='spearman')

# Create heatmaps
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', center=0,
            square=True, ax=ax1, fmt='.3f')
ax1.set_title('Pearson Correlation Heatmap')

# Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='YlGnBu', center=0,
            square=True, ax=ax2, fmt='.3f')
ax2.set_title('Spearman Correlation Heatmap')

plt.tight_layout()
plt.show()
```
